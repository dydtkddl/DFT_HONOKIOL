```python
from pymatgen.core import Structure, Lattice

# 원래 CIF 불러오기
structure = Structure.from_file("LAFRAN01_clean.cif")

# 현재 격자 정보 가져오기
lattice = structure.lattice
new_lattice = Lattice.from_lengths_and_angles(
    [lattice.a / 4, lattice.b / 4, lattice.c / 4],
    lattice.angles
)

# 새로운 격자로 구조 만들기 (좌표는 wrapping됨)
structure_64x = Structure(new_lattice, structure.species, structure.frac_coords, to_unit_cell=True)

# 저장
structure_64x.to(filename="volume_reduced_64x.cif")
```
